```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img = Image.open(r"../input/lukejpg/luke.jpg")
plt.imshow(img)
plt.axis("off")
plt.show()

img_array = np.array(img, dtype=np.float32)
img_array = np.expand_dims(img_array, axis=0)
print(img_array.shape)
```

```python
datagen = ImageDataGenerator(
    rotation_range=90,  # 旋转，旋转角度是在指定的范围内随机
#     shear_range=0.5,  # 剪切角度
#     shear_range=20.0,
#     horizontal_flip=True  # 水平翻转
#     vertical_flip=True,  # 垂直翻转
)

datagen.fit(img_array)
my_generator = datagen.flow(img_array, batch_size=1)
res = my_generator.next()
print(res.shape)

plt.imshow(np.int_(res[0]))
plt.axis("off")
plt.show()
```

```python
datagen = ImageDataGenerator(
#     rotation_range=20,
#     shear_range=50.0,
#     horizontal_flip=True  # 水平翻转
    vertical_flip=True,  # 垂直翻转
)

datagen.fit(img_array)
my_generator = datagen.flow(img_array, batch_size=1)
res = my_generator.next()
print(res.shape)

plt.imshow(np.int_(res[0]))
plt.axis("off")
plt.show()
```

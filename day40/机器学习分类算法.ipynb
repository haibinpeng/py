# 分类算法

## K近邻

```python
# K近邻预测用户签到位置

# 读取数据
data = pd.read_csv('./data/FBlocation/train.csv')
print(data.head(10))
print('\n')
print(data.shape)
print('\n')
# 处理数据
# 缩小数据，减少计算时间
data = data.query('x > 1.0 & x < 1.25 & y > 2.5 & y < 2.75')
# 处理时间的数据
time_value = pd.to_datetime(data['time'], unit='s')  # 秒
print(time_value)
print('\n')

# 把日期、时间格式转换为字典格式
time_value = pd.DatetimeIndex(time_value)
print(time_value)

# 构造一些特征
data['day'] = time_value.day
data['hour'] = time_value.hour
data['weekday'] = time_value.weekday
# 把time特征删除
data = data.drop(['time'], axis=1)
print(data)
print('\n')

# 每个地点去的人数
place_count = data.groupby('place_id').count()
print(place_count)
print('\n')
# 只选择去的人大于3的数据，认为1,2,3的是噪音，这个地方去的人很少
tf = place_count[place_count.row_id > 3].reset_index()
print(tf)

# 根据设定的地点目标值，对原本的样本进行过滤
data = data[data['place_id'].isin(tf.place_id)]
print(data)

# 取出处理后的数据当中的特征值和目标值
# y是目标值
y = data['place_id']
# print(y)
# 删除目标值，保留特征值
x = data.drop(['place_id'], axis=1)
# 删除无用的特征值,认为row_id是无用的
x = x.drop(['row_id'], axis=1)
print(x.shape)
print('\n')
print(x.columns)

# 进行数据的分割，训练集合测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
# 特征工程（标准化）
std = StandardScaler()
# 对测试集和训练集的特征值进行标准化，训练集用fit_transform
x_train = std.fit_transform(x_train)
print(std.mean_)
print(std.var_)
print('\n')
# 测试集用transform，transform不再进行均值和方差的计算
x_test = std.transform(x_test)
print(std.mean_)
print(std.var_)

# 进行算法流程
# 超参数，可以通过设置n_neighbors，来调整结果好坏
# 网格搜索和交叉验证
knn = KNeighborsClassifier()
param = {'n_neighbors': [1, 3, 5, 8, 10]}
gc = GridSearchCV(knn, param_grid=param,cv=3)
gc.fit(x_train, y_train)
y_predict = gc.predict(x_test)
print('预测的目标签到位置为：', y_predict)
print('预测的准确率：', gc.score(x_test, y_test))
print("在交叉验证当中最好的结果：", gc.best_score_)
print("选择最好的模型是：", gc.best_estimator_)
print("每个超参数每次交叉验证的结果：", gc.cv_results_)
```

## 朴素贝叶斯

```python
news = fetch_20newsgroups(subset='all', data_home='data')
print(len(news.data))
print(news.target)
print(news.target_names)
print('\n')

# 数据分割
# 训练集的特征值x_train 测试集的特征值x_test 训练集的目标值y_train 测试集的目标值y_test
x_train, x_test, y_train, y_test = train_test_split(news.data, news.target, test_size=0.25, random_state=1)

# 特征工程
tf = TfidfVectorizer()
x_train = tf.fit_transform(x_train)
x_test = tf.transform(x_test)

# 机器学习（模型训练）
# 朴素贝叶斯算法,alpha是拉普拉斯平滑系数
mlt = MultinomialNB(alpha=1.0)
mlt.fit(x_train, y_train)

# 模型评估
y_predict = mlt.predict(x_test)
print('预测的文章类别为：', y_predict)
print('\n')
print('准确率为：', mlt.score(x_test, y_test))
print('\n')
print('每个类别的精确率和召回率：', classification_report(y_test, y_predict, target_names=news.target_names))
```

### AUC

```python
# 把0-20总计20个分类变为0和1
y_test1 = np.where(y_test == 5, 1, 0)
print(y_test1.sum())  # 测试集的目标值中为第5类的数量
y_predict1 = np.where(y_predict == 5, 1, 0)
print(y_predict1.sum())  # 预测的为第五类的数量
# roc_auc_score的y_test只能是二分类
print('AUC指标：', roc_auc_score(y_test1, y_predict1))
```

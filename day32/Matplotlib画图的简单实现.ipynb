```python
# 我是注释
import matplotlib.pyplot as plt
# 传入x和y，通过plot画图
plt.plot([1, 0, 9], [4, 5, 6])
# 在执行程序的时候展示图形
plt.show()
```

```python
# x轴的位置
x = range(1,8)
y = [17, 17, 18, 15, 11, 11, 13]
# 传入x和y, 通过plot画折线图
plt.plot(x,y,color='red',alpha=0.2,linestyle=':',linewidth=3)
plt.show()
```

```python
#折点样式
# x轴的位置
x = range(1,8)
y = [17, 17, 18, 15, 11, 11, 13]
# 传入x和y, 通过plot画折线图
plt.plot(x, y, marker='o')
plt.show()
```
